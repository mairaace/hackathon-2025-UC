```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
import pickle
import json
```

```python
df_cursos = pd.read_csv("data/Coursera_courses.csv")
df_cursos.head()
```

Tabla de las personas con sus ids: 

```python
df_nombre_id = df_cursos[['name', 'course_id']]
df_nombre_id.head()
```

```python
df_nombre_id.to_csv('data/nombre_id_cursos.csv', index=False)
```

```python
df_resenas = pd.read_csv("data/Coursera_reviews.csv")
df_resenas.head(20)
```

```python
conteo_por_usuario = df_resenas['reviewers'].value_counts()
top_300_usuarios = conteo_por_usuario.head(3000).index
df_resenas_reducido = df_resenas[df_resenas['reviewers'].isin(top_300_usuarios)]
```

Tabla de calificaciones por persona:

```python
tabla_calificaciones = df_resenas_reducido.pivot_table(
    index='reviewers',     
    columns='course_id',   
    values='rating',       
    aggfunc='mean'
).fillna(0).astype(int).reset_index()
```

```python
tabla_calificaciones.head()
```

```python
df_numerico = tabla_calificaciones.set_index('reviewers')
conteo_ratings = (df_numerico != 0).sum()
top_cursos = conteo_ratings.sort_values(ascending=False).head(295).index
```

```python
columnas_finales = ['reviewers'] + list(top_cursos)
tabla_filtrada = tabla_calificaciones[columnas_finales]
```

```python
tabla_filtrada.head()
```

```python
tabla_filtrada.to_csv('data/calificacion_cursos.csv', index=False)
```

```python
tabla_final = tabla_filtrada.drop(columns=['reviewers'])
tabla_final.head()
```

Sistemas de recomendacin que utilizan KNN

```python
matriz_final = tabla_final.values
matriz_final
```

```python
mean_user_rating = np.mean(matriz_final, axis=1).reshape(-1, 1)
matriz_normalizada = matriz_final - mean_user_rating
```

```python
knn = NearestNeighbors(metric='euclidean', algorithm='brute', n_neighbors=5)
knn.fit(matriz_normalizada)
```

```python
target_user_index = 0
distances, indices = knn.kneighbors(matriz_normalizada[target_user_index].reshape(1, -1), n_neighbors=3)
```

```python
neighbors_ratings = matriz_final[indices.flatten()]
# Promediar solo valores no-cero que son las calificasiones que si se han echo
predicted_ratings = np.zeros(neighbors_ratings.shape[1])
for i in range(neighbors_ratings.shape[1]):
    non_zero = neighbors_ratings[:, i][neighbors_ratings[:, i] != 0]
    if len(non_zero) > 0:
        predicted_ratings[i] = non_zero.mean()
    else:
        predicted_ratings[i] = 0
```

```python
unrated_items = np.where(matriz_final[target_user_index] == 0)[0]
recommended_items = unrated_items[np.argsort(predicted_ratings[unrated_items])[::-1]]

print(f"Recommended items for user {target_user_index}: {recommended_items}")
```

```python
# Crear mapeo de nombre de usuario a ndice
usuarios_lista = tabla_filtrada['reviewers'].values
usuario_a_indice = {usuario: idx for idx, usuario in enumerate(usuarios_lista)}

# nomrbe de ususaro
nombre_usuario = input("Ingresa el nombre del usuario: ")

#validar nombre de usuario
if nombre_usuario not in usuario_a_indice:
    print(f"Usuario '{nombre_usuario}' no encontrado")
else:
    target_user_index = usuario_a_indice[nombre_usuario]  
    
    # Entrenar KNN para este usuario
    distances, indices = knn.kneighbors(matriz_normalizada[target_user_index].reshape(1, -1), n_neighbors=3)
    
    neighbors_ratings = matriz_final[indices.flatten()]
    predicted_ratings = neighbors_ratings.mean(axis=0)
    
    # Obtener items sin calificar
    unrated_items = np.where(matriz_final[target_user_index] == 0)[0]
    predicted_unrated = predicted_ratings[unrated_items]
    sorted_indices = np.argsort(predicted_unrated)[::-1]
    top_10_indices = sorted_indices[:10]
    top_10_items = unrated_items[top_10_indices]
    
    # Mostrar top 10 con nombres
    print(f"\n Top 10 cursos recomendados para {nombre_usuario}:\n")
    for rank, item_idx in enumerate(top_10_items, 1):
        course_id = list(top_cursos)[item_idx]
        course_name = df_cursos[df_cursos['course_id'] == course_id]['name'].values[0]
        rating_predicho = predicted_ratings[item_idx]
        print(f"{rank}. {course_name}")
        print(f"   Calificacin predicha: {rating_predicho:.2f}\n")
```

```python

# Guardar el modelo KNN
with open('data_web/modelo_knn.pkl', 'wb') as f:
    pickle.dump(knn, f)

# Guardar la matriz normalizada
with open('data_web/matriz_normalizada.pkl', 'wb') as f:
    pickle.dump(matriz_normalizada, f)

# Guardar la matriz final
with open('data_web/matriz_final.pkl', 'wb') as f:
    pickle.dump(matriz_final, f)

# Guardar datos en JSON
datos = {
    "usuarios": tabla_filtrada['reviewers'].tolist(),
    "cursos": list(top_cursos),
    "nombres_cursos": df_cursos.set_index('course_id')['name'].to_dict()
}

with open('data_web/datos_recomendador.json', 'w') as f:
    json.dump(datos, f)

```
