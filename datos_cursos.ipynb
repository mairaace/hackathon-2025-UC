```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
```

```python
df_cursos = pd.read_csv("data/Coursera_courses.csv")
df_cursos.head()
```

Tabla de las personas con sus ids: 

```python
df_nombre_id = df_cursos[['name', 'course_id']]
df_nombre_id.head()
```

```python
df_nombre_id.to_csv('data/nombre_id_cursos.csv', index=False)
```

```python
df_resenas = pd.read_csv("data/Coursera_reviews.csv")
df_resenas.head(20)
```

```python
conteo_por_usuario = df_resenas['reviewers'].value_counts()
top_300_usuarios = conteo_por_usuario.head(3000).index
df_resenas_reducido = df_resenas[df_resenas['reviewers'].isin(top_300_usuarios)]
```

Tabla de calificaciones por persona:

```python
tabla_calificaciones = df_resenas_reducido.pivot_table(
    index='reviewers',     
    columns='course_id',   
    values='rating',       
    aggfunc='mean'
).fillna(0).astype(int).reset_index()
```

```python
tabla_calificaciones.head()
```

```python
df_numerico = tabla_calificaciones.set_index('reviewers')
conteo_ratings = (df_numerico != 0).sum()
top_cursos = conteo_ratings.sort_values(ascending=False).head(295).index
```

```python
columnas_finales = ['reviewers'] + list(top_cursos)
tabla_filtrada = tabla_calificaciones[columnas_finales]
```

```python
tabla_filtrada.head()
```

```python
tabla_filtrada.to_csv('data/calificacion_cursos.csv', index=False)
```

```python
tabla_final = tabla_filtrada.drop(columns=['reviewers'])
tabla_final.head()
```

Sistemas de recomendación que utilizan KNN

```python
matriz_final = tabla_final.values
matriz_final
```

```python
mean_user_rating = np.mean(matriz_final, axis=1).reshape(-1, 1)
matriz_normalizada = matriz_final - mean_user_rating
```

```python
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(matriz_normalizada)
```

```python
target_user_index = 0
distances, indices = knn.kneighbors(matriz_normalizada[target_user_index].reshape(1, -1), n_neighbors=3)
```

```python
neighbors_ratings = matriz_final[indices.flatten()]
predicted_ratings = neighbors_ratings.mean(axis=0)
```

```python
unrated_items = np.where(matriz_final[target_user_index] == 0)[0]
recommended_items = unrated_items[np.argsort(predicted_ratings[unrated_items])[::-1]]

print(f"Recommended items for user {target_user_index}: {recommended_items}")
```
